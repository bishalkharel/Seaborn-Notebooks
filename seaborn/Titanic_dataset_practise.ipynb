### In this notebook we are going to practise the titanic dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from PIL import Image
```

```python
df=pd.read_csv("titanic_dataset.csv")
df
```

```python

sns.heatmap(df.isnull())
```

lets count the survivors in according to sex 

```python
sns.countplot("Survived",data=df)
```

```python

plt.figure(figsize=(9,9),dpi=500)
sns.countplot(x="Survived",data=df,hue="Sex")
```

```python
# from the given plot we can see that in passenger class no 3 has died mostly 
sns.countplot(x="Survived",data=df,hue="Pclass")
```

```python
plt.figure(figsize=(9,9),dpi=300)
sns.distplot(df["Age"].dropna(),kde=False,color="darkred",bins=40)
```

```python
sns.countplot("SibSp",data=df)
```

So we have a lot a NaN values in age rows and cabin so we have to replace those values. Instead of removing age or cabin features or columns we can put the average age of the people. To see or visualize the values we will use box plot

```python
sns.boxplot(y="Pclass",x="Age",data=df,palette="winter")
```
