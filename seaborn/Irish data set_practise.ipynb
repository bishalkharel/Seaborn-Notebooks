## In this notebook we are going to practise all the plots on irish dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("iris")
df.head()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
# joint plot
sns.jointplot("sepal_length","petal_length",data=df,kind="hex")
```

```python
# pairplot

sns.pairplot(df,hue="sepal_length")
```

```python
# distplot
sns.distplot(df["sepal_length"])
```

## Categorical plot

```python
# count plot
sns.countplot("species",data=df)
```

```python
# box plot

sns.boxplot("sepal_length","sepal_width",data=df)
```

```python
sns.barplot(x="species",y="petal_length",data=df)
```

```python
plt.figure(figsize=(9,8),dpi=400)
sns.violinplot(x="petal_length",y="species",data=df)
```
